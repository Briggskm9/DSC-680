```python
#Assign X as the original ratings dataframe and y as the user_id column of ratings.
X = df.copy()
y = df["kp"]
#Split into training and test datasets, stratified along user_id
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, stratify=y, random_state=42)
```

```python
#sparse dataframe

df_r = X_train.pivot(columns='predicted_time', values='kp')
```

```python
df_r
```

```python
df_rate = df_r.copy().fillna(0)
df_rate.head()
```

```python
#cosine similarity 
similarity_matrix = cosine_similarity(df_rate, df_rate)
similarity_matrix_df = pd.DataFrame(similarity_matrix, index=df_rate.index, columns=df_rate.index)
```

```python
#calculate ratings using weighted sum of cosine similarity
#function to calculate ratings

def calculate_ratings(predicted_time, kp):
    if predicted_time in df_rate:
        cosine_scores = similarity_matrix_df[kp] #similarity of predicted date with every other user
        ratings_scores = df_rate[predicted_time]      #ratings of every other user for the kp
        index_not_rated = ratings_scores[ratings_scores.isnull()].index
        ratings_scores = ratings_scores.dropna()
        cosine_scores = cosine_scores.drop(index_not_rated)
#calculating rating by weighted mean of ratings and cosine scores of the users who have rated the movie
        rate = np.dot(ratings_scores, cosine_scores)/cosine_scores.sum()
        return rate
```

```python
calculate_ratings(3,2)
```

```python
#evaluates on test set
def score_on_test_set():
    rate_pairs = zip(X_test['predicted_time'], X_test['kp'])
    predicted_ratings = np.array([calculate_ratings('predicted_time', 'kp') for (predicted_time, kp) in rate_pairs])
    true_ratings = np.array(X_test[df])
    score = np.sqrt(mean_squared_error(true_ratings, predicted_ratings))
    return score
test_set_score = score_on_test_set()
print(test_set_score)
```
